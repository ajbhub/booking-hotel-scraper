```python
import dateutil.parser as dparser
import pandas as pd
import re
from datetime import datetime
import hvplot.pandas
from datetime import date, timedelta

pd.options.plotting.backend = 'holoviews'
```

```python
def date_range_list(start_date, end_date):
    # Return list of datetime.date objects between start_date and end_date (inclusive).
    date_list = []
    curr_date = start_date
    while curr_date <= end_date:
        date_list.append(curr_date)
        curr_date += timedelta(months=1)
    return date_list
```

```python
df1 = pd.read_csv('2022-05-01 booking data.csv')
df2 = pd.read_csv('2022-06-01 booking data.csv')

combined = df1.append(df2)
combined = combined.drop(columns = ["Unnamed: 0"])
combined.set_index('date', inplace=True)



df = combined.pivot(columns='name', values='price')

#dflimited = df[['[\'Adrift Hotel\']', '[\'Astoria Riverwalk Inn\']', '[\'Boardwalk Cottages\']', '[\'The Breakers Long Beach\']', '[\'Chautauqua Resort & Conference Center\']']].copy()
dfadrift = df[['[\'Adrift Hotel\']', '[\'Chautauqua Resort & Conference Center\']', '[\'Boardwalk Cottages\']']].copy()


#print(dflimited.head())

fig = dfadrift.hvplot(width=1800, height = 600, rot=90)

display(fig)
```
